```python
import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/canada_per_capita_income.csv')
df
```

```python
%matplotlib inline
plt.xlabel('year')
plt.ylabel('income')
plt.scatter(df.year,df.income)
```

```python
new_year = df.drop('income',axis = 'columns')
new_year
```

```python
income = df.income
income
```

```python
reg = linear_model.LinearRegression()
reg.fit(new_year,income)
```

```python
reg.predict([[2050]])
```
